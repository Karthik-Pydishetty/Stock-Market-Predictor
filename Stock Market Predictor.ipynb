```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
#ticker class which enables to download price history for a single symbol
```

```python
sp500 = sp500.history(period="max")
#query all data from the vv beg when the index was created
```

```python
sp500 = sp500.iloc[:, :-2]
sp500
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) # shifts close column one row down
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) 
# boolean value if tomorrows price is higher than yesterdays 
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() # avoid copy warning
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, min_samples_split=100, random_state=1) 
# n_estimators are the num of trees to train, random state to have reproduceable results

train = sp500.iloc[:-100]

test = sp500.iloc[-100:]

predictors = ["Close", "Open", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])

```

```python
from sklearn.metrics import precision_score

# error metric

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1) # treat each input as a column in dataset
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors]) # generating predictions
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1) # df with predictions and actual target values
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250): # take first 10 years of data and then predict values for the eleventh year
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy() # df is defined as the beggining to the ith row
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
        
        
        
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# join data frames together
vix = yf.download('^VIX', period='max')['Close']
if sp500.index.tz is not None:
    sp500.index = sp500.index.tz_convert(None)
if vix.index.tz is not None:
    vix.index = vix.index.tz_convert(None)

vix.name = 'VIX_Close'
vix_aligned = vix.reindex(sp500.index, method='ffill')
sp500 = pd.concat([sp500, vix_aligned], axis=1).dropna()


```

```python
# more predictors
# rolling averages
# rolling correlation with VIX
horizons = [2,5,60,250,1000] # last 2 days, week, year, 
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"] # sum of previous column, how many days did it actually go up
    new_predictors += [ratio_column, trend_column]
    
for horizon in horizons:
    corr_column = f"Rolling_Correlation_{horizon}"
    sp500[corr_column] = sp500['Close'].rolling(horizon).corr(sp500['VIX_Close'])
    new_predictors.append(corr_column)

```

```python
sp500 = sp500.dropna() #drop missing rows
sp500
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split=200, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors]) [:,1] # generating predictions
    preds[preds >= .6] = 1 # model has to be more confident when predicting price will go up
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1) # df with predictions and actual target values
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
